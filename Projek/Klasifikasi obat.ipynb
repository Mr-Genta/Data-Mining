```python
import pandas as pd
```

```python
obat = pd.read_csv('drug200.csv', sep=';')
```

```python
data = obat.copy()
```

```python
target = 'Drug'
```

```python
encode = ['Jenis Kelamin','tekanan darah','Cholesterol']
```

```python
data
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col],prefix=col)
    data = pd.concat([data,dummy], axis=1)
    del data[col]
```

```python
data
```

```python
target_mapper ={'drugA':0,'drugB':1,'drugC':2,'drugX':3,'drugY':4,}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['Drug'] = data['Drug'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('Drug',axis=1)
Y = data['Drug']
```

```python
X
```

```python
Y
```

```python
# Model NBC
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X,Y)
```

```python
# Menyimpan Model NBC
import pickle
pickle.dump(model, open('modelNBC_obat.pkl', 'wb'))
```
